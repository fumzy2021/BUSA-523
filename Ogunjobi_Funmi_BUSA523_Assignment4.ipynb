Using the dataset provided, develop the following visualizations using the right
set of variables

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data_sheet = pd.read_csv('data.csv', header = 0, index_col = 0)

data_sheet
```

a) Histogram

```python
plt.hist(data_sheet['Temperature'])
plt.suptitle('Histogram of Temperature')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
```

b) Barchart

```python
dataBar = pd.DataFrame(data_sheet, columns=['Unemployment'])
dataBar[0:143:25].plot.bar()
plt.suptitle('Unemployement rate over 3 years')
plt.ylabel('Unemployment rate')
```

c) Line Chart

```python
dataLine = pd.DataFrame(data_sheet, columns=['Fuel_Price'])
dataLine.plot()
plt.suptitle('Fuel Prices over the years')
plt.ylabel('Fuel Price per gallon in USD')
```

d)Area Chart

```python
dataArea = pd.DataFrame(data_sheet, columns=['dept_1 sales','dept_2 sales','dept_3 sales'])
dataArea.plot.area()
plt.suptitle('Sales by Department from 2010 to 2012')
plt.ylabel('Sales')
```

e) ScatterPlot

```python
dataScatter = pd.DataFrame(data_sheet)
dataScatter.plot.scatter(x = 'CPI', y = 'Fuel_Price')
plt.suptitle('How CPI affects Fuel Prices')
```
